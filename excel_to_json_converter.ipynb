```python
import json

with open("/content/Crct_Use.json") as f1, open("/content/wrong_user.json") as f2:
    data1 = json.load(f1)
    data2 = json.load(f2)

ids1 = {emp["EmployeeId"] for emp in data1}
ids2 = {emp["EmployeeId"] for emp in data2}

only_in_file1 = [emp for emp in data1 if emp["EmployeeId"] not in ids2]
only_in_file2 = [emp for emp in data2 if emp["EmployeeId"] not in ids1]

# ✅ structured result
result = {
    "only_in_file1": only_in_file1,
    "only_in_file2": only_in_file2
}

# 👉 Pretty print (valid JSON format)
print(json.dumps(result, indent=2))

# 👉 Save to file (recommended)
with open("/content/difference.json", "w") as out:
    json.dump(result, out, indent=2)
```

```python
import json

with open("/content/Crct_Use.json") as f1, open("/content/wrong_user.json") as f2:
    data1 = json.load(f1)
    data2 = json.load(f2)

ids1 = {emp["EmployeeId"] for emp in data1}
ids2 = {emp["EmployeeId"] for emp in data2}

only_in_file1 = [emp for emp in data1 if emp["EmployeeId"] not in ids2]
only_in_file2 = [emp for emp in data2 if emp["EmployeeId"] not in ids1]

# ✅ structured result with counts
result = {
    "file1_count": len(data1),
    "file2_count": len(data2),
    "only_in_file1_count": len(only_in_file1),
    "only_in_file2_count": len(only_in_file2),
    "only_in_file1": only_in_file1,
    "only_in_file2": only_in_file2
}

# 👉 Pretty print
print(json.dumps(result, indent=2))

# 👉 Save to file
with open("/content/difference.json", "w") as out:
    json.dump(result, out, indent=2)
```
